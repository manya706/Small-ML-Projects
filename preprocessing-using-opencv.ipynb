```python
# import required libraries
import cv2
from google.colab.patches import cv2_imshow
# load the input image
img = cv2.imread('/content/demo image.jfif')

# convert the input image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# apply thresholding to convert grayscale to binary image
ret,thresh = cv2.threshold(gray,70,255,0)
ret, thresh2 = cv2.threshold(gray,70,255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
thresh3 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, 199, 5)
# Display the Binary Image
print('original image:')
cv2_imshow(img)
print('simple thresholding:')
cv2_imshow(thresh)
print('Otsu thresholding:')
cv2_imshow(thresh2)
print('Adaptive thresholding:')
cv2_imshow(thresh3)
```

```python
# Importing numpy
import numpy as np
# Importing matplotlib.pyplot
import matplotlib.pyplot as plt
# Averaging the image
img1 = cv2.blur(img, (3,3))
# Displaying the blurred image
plt.imshow(img1)
```

```python
img2 = cv2.GaussianBlur(img, (5,5), cv2.BORDER_DEFAULT)
# Displaying the blurred image
plt.imshow(img2)
```

```python
img3 = cv2.medianBlur(img, 1)
# Displaying the blurred image
plt.imshow(img3)
```

```python
# Bilateral Blur
img4 = cv2.bilateralFilter(img,9, 75, 75)
plt.imshow(img4)
```

```python
# Displaying the original BGR image
cv2_imshow(img)

# Using cv2.split() to split channels of coloured image
b,g,r = cv2.split(img)

# Displaying Blue channel image
# Blue colour is highlighted the most
cv2_imshow( b)

# Displaying Green channel image
# Green colour is highlighted the most
cv2_imshow(g)

# Displaying Red channel image
# Red colour is highlighted the most
cv2_imshow(r)
```

```python
if img is None:
    print("Error: Could not load image.")
else:
    # Convert the color image to grayscale
    grayscale_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Define the intensity value for brightness adjustment
    intensity_value = 70

    # Increase brightness by adding the intensity value to all pixels
    brighter_image = cv2.add(grayscale_image, intensity_value)

    # Decrease brightness by subtracting the intensity value from all pixels
    darker_image = cv2.subtract(grayscale_image, intensity_value)

    # Display the original, brighter, and darker images
    cv2_imshow(grayscale_image)
    cv2_imshow(brighter_image)
    cv2_imshow(darker_image)
```

```python
image = img
if image is None:
    print("Error: Could not load image.")
else:
    # Convert the color image to grayscale
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Threshold the grayscale image to create a binary image
    _, binary_image = cv2.threshold(grayscale_image, 128, 255, cv2.THRESH_BINARY)

    # Create a black image with a white rectangle and circle
    black_image = np.zeros_like(binary_image)
    cv2.rectangle(black_image, (50, 50), (150, 150), 255, -1)  # White rectangle
    cv2.circle(black_image, (100, 100), 30, 255, -1)  # White circle

    # Perform bitwise AND operation
    bitwise_and = cv2.bitwise_and(binary_image, black_image)

    # Perform bitwise OR operation
    bitwise_or = cv2.bitwise_or(binary_image, black_image)

    # Perform bitwise XOR operation
    bitwise_xor = cv2.bitwise_xor(binary_image, black_image)

    # Perform bitwise NOT operation on the binary image
    bitwise_not = cv2.bitwise_not(binary_image)

    # Display the original binary image and the results of bitwise operations
    cv2_imshow(binary_image)
    cv2_imshow(bitwise_and)
    cv2_imshow(bitwise_or)
    cv2_imshow(bitwise_xor)
    cv2_imshow(bitwise_not)
```
